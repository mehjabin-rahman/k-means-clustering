```python
# def activation function

def activation(out, threshold):
    if out > threshold:
        return 1
    else:
        return 0
    
    
# def perceptron/single layer Neural Network
def single_layer_NN(or_input):
    a = [0,0,1,1]
    b = [0,1,0,1]
    y = [0,1,1,1]
    w = [0.2,0.3]
    
    threshold = 0.1
    lr = 0.1
    
    i = 0
    
    print('Perceptron Training: ')
    print('**********************')
    print("____________________")
    
    while i < 4:
        summation = a[i]*w[0] + b[i]*w[1]
        o = activation(summation, threshold)
        print("Input: " + str(a[i]) + " , " + str(b[i]))
        print("Weights: " + str(w[0]) + " , " + str(w[1]))
        print("Summation: " + str(summation) + " , " + str(threshold))
        print("Actual Output: " + str(y[i]) + " Predicted Output: " + str(o))
        
        if (o != y[i]):
        # w = w + lr(actual output - predicted output) * input
            print("________\n Updating Weights")
            w[0] = w[0] + lr*(y[i]-o)*a[i]
            w[1] = w[1] + lr*(y[i]-o)*a[i]
        
            print("Updated Wrights" + str(w[0]) + " , " + str(w[1]))
            i = -1
            print("\nWeights updated training again: ")
            print("*****************************")
        
    
        i += 1
    
        print("____________")
    
    
    # OR Gate function
    summation = or_input[0]*w[0] + or_input[1]*w[1]

    return activation(summation, threshold)

T1 = [0,1]
T2 = [1,0]
T3 = [1,1]
T4 = [0,0]
print("OR Gate Output For " + str(T1) + " : " + str(single_layer_NN(T1)))
print("OR Gate Output For " + str(T2) + " : " + str(single_layer_NN(T2)))
print("OR Gate Output For " + str(T3) + " : " + str(single_layer_NN(T3)))
print("OR Gate Output For " + str(T4) + " : " + str(single_layer_NN(T4)))

        
```
